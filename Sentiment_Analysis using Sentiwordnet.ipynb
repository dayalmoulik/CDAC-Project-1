```python
import pandas as pd
import numpy as np
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
import re
import spacy
from nltk.corpus import wordnet as wn
from nltk.corpus import sentiwordnet as swn
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df=pd.read_csv('preprocessed_dataset/SarTanSeJuda.csv')
df.head()
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
df['lemmatized_text']=df['preprocessed_text'].apply(lambda x: [lemmatizer.lemmatize(w) for w in x.split()])
```

```python
df.head()
```

```python
postagging = []

for review in df['lemmatized_text']:
    postagging.append(nltk.pos_tag(review))

df['pos_tags'] = postagging
```

```python
df.head()
```

```python
def penn_to_wn(tag):
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    return None
```

```python
def get_sentiment(word,tag):
    wn_tag = penn_to_wn(tag)
    
    if wn_tag not in (wn.NOUN, wn.ADJ, wn.ADV):
        return []

    #Lemmatization
    lemma = lemmatizer.lemmatize(word, pos=wn_tag)
    if not lemma:
        return []

    synsets = wn.synsets(word, pos=wn_tag)
    if not synsets:
        return []

    # Take the first sense, the most common
    synset = synsets[0]
    swn_synset = swn.senti_synset(synset.name())

    return [synset.name(), swn_synset.pos_score(),swn_synset.neg_score(),swn_synset.obj_score()]


pos=neg=obj=count=0
    
    
senti_score = []

for pos_val in df['pos_tags']:
    senti_val = [get_sentiment(x,y) for (x,y) in pos_val]
    for score in senti_val:
        try:
            pos = pos + score[1]  
        except:
            continue
    senti_score.append(pos - neg)
    pos=neg=0    
    
df['senti_score'] = senti_score
print(df['senti_score'])

print(df.head)
```

```python
overall=[]
for i in range(len(df)):
    if df['senti_score'][i]>= 0.05:
        overall.append('Positive')
    elif df['senti_score'][i]<= -0.05:
        overall.append('Negative')
    else:
        overall.append('Neutral')
df['Overall Sentiment']=overall
```
